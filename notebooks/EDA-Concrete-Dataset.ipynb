### Observe the distribution of the Label

```python
plt.figure(figsize=(13, 15))
plt.hist(df_concrete['compressive_strength'], bins=20, color='maroon')
plt.title("Distribution of the Compressive Strength")
plt.xlabel("Strength (MPa)")
plt.ylabel("Frequency")
plt.show()
```

### Identify a pattern and establish a relationship between 'Compressive Strength' (Label) and other features

### 1.Quantity of Cement where the Strength is greater than mean value

```python
high_strength = df_concrete[df_concrete['compressive_strength'] >
                            df_concrete['compressive_strength'].mean()]

df = high_strength[['cement', 'compressive_strength']].sort_values(by='compressive_strength', ascending=False)
df
```

### Bar Plot 

```python
x = df['compressive_strength'].iloc[:150]
y = df['cement'].iloc[:150]
title = "Cement vs High Compressive Strength"
x_label = "Compressive Strength (MPa)"
y_label = "Cement (kg in a m^3 mixture)"

plt.figure(figsize=(13, 15))
plt.bar(x, y)

plt.title(title, fontsize=16)
plt.xlabel(x_label)
plt.xticks()
plt.ylabel(y_label)
plt.show()
```

### Covariance Matrix between Cement and High Compressive Strength

```python
title = "Covariance Matrix -Cement vs High Compressive Strength"

corr = df.select_dtypes('number').corr()
sb.heatmap(corr, cmap="Blues", annot=True)
plt.title(title)
plt.show()
```
